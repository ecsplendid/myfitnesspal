```python
from pandas import read_csv
from pipe import *
from dfply import *
import matplotlib.pyplot as plt

df = (read_csv('Nutrition.csv') >>
 mask(X.Date > '2016-01-01') >>
 arrange(X.Date) >>
 group_by(X.Date) 
)

df.columns = ['Date', 'Meal', 'Calories', 'Fat', 'SaturatedFat',
       'PolyunsaturatedFat', 'MonounsaturatedFat', 'TransFat',
       'Cholesterol', 'Sodium', 'Potassium', 'Carbohydrates', 'Fiber',
       'Sugar', 'Protein', 'VitaminA', 'VitaminC', 'Calcium', 'Iron']

df >> summarize(Protein=X.Protein.sum()) 
```
